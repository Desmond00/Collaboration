```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv("kddTrain.csv")
trainData = data.drop('41', axis=1)
```

```python
trainData.head()
```

```python
''' As the column no.s 19 and 20 do not have any change in value so they are not contributing in S.D. and mean score
which is required in Pearson Corelation so we are not including them.
For more info see https://stackoverflow.com/questions/22655667/dataframe-correlation-produces-nan-although-its-values-are-all-integers
'''
```

```python
data.drop('19', axis=1, inplace=True)
data.drop('20', axis=1, inplace=True)
```

```python
scoreTable = data.corr(method='pearson')
```

```python
data.corr(method='pearson').style.format("{:.2}").background_gradient(cmap=plt.get_cmap('coolwarm'), axis=1)
```

```python
type(scoreTable)
```

```python
FeatureIndices = scoreTable.index.tolist()
```

```python
def getContinuousFeatureList() :
    i = 0
    j = 0
    ContinuousValuedFeatures = []
    while(i<41):
        j=0
        while(j<len(trainData)):
            FeatureValue = trainData.iloc[j][i]
            if(FeatureValue != 0 and FeatureValue != 1):
                ContinuousValuedFeatures.append(i)
                print('value: ',FeatureValue,'coordinates : ',j,i)
                j = len(trainData)
            j+=1
        i+=1
    return ContinuousValuedFeatures
```

```python
ContinuousValuedFeatures = getContinuousFeatureList()
```

```python
allFeatures = list(map(int, scoreTable.index.tolist()))
```

```python
DiscreteValuedFeatures = [feature for feature in allFeatures if feature not in ContinuousValuedFeatures]
```

```python
DiscreteValuedFeatures
```

```python
ContinuousValuedFeatures
```

```python
AverageCorelationScores = {}
TotalCorelationScores = {}
DependantFeatures = {}

# j = 0
for i in FeatureIndices:
    sumOfFeatureScores = 0
    FeaturesConsidered = []
    for j in FeatureIndices:
        score = scoreTable.loc[i][j]
        if (score > 0.4) and (i!=j):
            FeaturesConsidered.append(j)
            sumOfFeatureScores += score
    DependantFeatures[int(i)] = FeaturesConsidered
    TotalCorelationScores[int(i)] = sumOfFeatureScores
    if len(FeaturesConsidered) > 0:
        AverageCorelationScores[int(i)] = sumOfFeatureScores/len(FeaturesConsidered)
    else:
        AverageCorelationScores[int(i)] = 0
```

```python
for feature,score in AverageCorelationScores.items():
    print('feature : ',feature,'\tAverageCorelationScore : ',score,'\tTotalCorelationScore : ',TotalCorelationScores[feature],'\tDependantFeatures : ',DependantFeatures[feature])
```

```python
ContinuousValuedFeatureScores = {}
for i in ContinuousValuedFeatures:
    ContinuousValuedFeatureScores[i] = TotalCorelationScores[i]
```

```python
DiscreteValuedFeatureScores = {}
for i in DiscreteValuedFeatures:
    DiscreteValuedFeatureScores[i] = TotalCorelationScores[i]
```

```python
for feature,score in ContinuousValuedFeatureScores.items():
    print(feature, score)
```

```python
def sortDic(UnsortedDictionary):
    dicItems = [(value,key) for key,value in UnsortedDictionary.items()]
    dicItems.sort()
    dicItems.reverse()
    dicItems = {key:value for value,key in dicItems}
    return dicItems
```

```python
ContinuousValuedFeatureScores = sortDic(ContinuousValuedFeatureScores)
DiscreteValuedFeatureScores = sortDic(DiscreteValuedFeatureScores)
print(ContinuousValuedFeatureScores)
print(DiscreteValuedFeatureScores)
```
