```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv("kddTrain.csv")
trainData = data.drop('41', axis=1)
```

```python
scoreTable = data.corr(method='pearson')
```

```python
data.corr(method='pearson').style.format("{:.2}").background_gradient(cmap=plt.get_cmap('coolwarm'), axis=1)
```

```python
type(scoreTable)
```

```python
AverageCorelationScores = []
DependantFeatures = []

i,j = 0,0
while i<len(data.columns):
    sumOfFeatureScores = 0
    FeaturesConsidered = []
    j = 0
    while j<len(data.columns):
        if scoreTable.iloc[i,j] > 0.4:
            FeaturesConsidered.append(j)
            sumOfFeatureScores += scoreTable.iloc[i,j]
        j += 1
    DependantFeatures.append(FeaturesConsidered)
    if len(FeaturesConsidered) > 0:
        AverageCorelationScores.append(sumOfFeatureScores/len(FeaturesConsidered))
    else:
        AverageCorelationScores.append(0)
    i += 1
```

```python
i=0
while i < len(AverageCorelationScores):
    print(AverageCorelationScores[i], DependantFeatures[i])
    i += 1
```
