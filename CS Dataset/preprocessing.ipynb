```python
import pandas as pd
```

```python
import numpy as np
```

```python
from sklearn.preprocessing import Imputer
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
trainingData = pd.read_csv('dataTrain.csv')
testingData = pd.read_csv('dataTest.csv')
```

```python
trainingData.head()
```

```python
testingData.head()
```

```python
trainingData.replace(to_replace='na', value=np.NaN, inplace = True)
```

```python
testingData.replace(to_replace='na', value=np.NaN, inplace = True)
```

```python
trainingData.head()
```

```python
testingData.head()
```

```python
def imputeMissingValues(data):
    values = data.drop('class', axis=1).values
    imputer = Imputer()
    transformed_values = imputer.fit_transform(values)
    print(np.isnan(transformed_values).sum())
    return pd.concat([pd.DataFrame(data=transformed_values, columns = data.drop('class', axis=1).columns), data.loc[:,'class']], axis=1)
```

```python
trainingData = imputeMissingValues(trainingData)
```

```python
testingData = imputeMissingValues(testingData)
```

```python
def minMaxScaling(data):
    y = data.loc[:,'class']
    data.drop('class', axis=1, inplace=True)
    scaler = MinMaxScaler()
    scaler.fit(data)
    return pd.concat([pd.DataFrame(data=scaler.transform(data), columns = data.columns), y], axis=1)
```

```python
trainingData = minMaxScaling(trainingData)
```

```python
testingData = minMaxScaling(testingData)
```
