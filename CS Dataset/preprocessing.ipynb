```python
import pandas as pd
```

```python
import numpy as np
```

```python
from sklearn.preprocessing import Imputer
```

```python
data = pd.read_csv('postures.csv')
```

```python
data.replace(to_replace='?', value=np.NaN, inplace = True)
```

```python
data.head()
```

```python
values = data.values
imputer = Imputer()
transformed_values = imputer.fit_transform(values)
```

```python
print(np.isnan(transformed_values).sum())
```

```python
data = pd.DataFrame(data=transformed_values, columns = data.columns)
```

```python
data
```

```python
data.to_csv('preprocessedData.csv', sep=',')
```
