```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv('preprocessed.csv')
```

```python
minMaxScaler = MinMaxScaler()
```

```python
minMaxScaler.fit(data)
ScaledData = minMaxScaler.transform(data)
```

```python
processedData = pd.DataFrame(data=ScaledData, columns = data.columns)
```

```python
processedData.head()
```

```python
processedData.to_csv('MinMaxScaledData.csv', sep=',')
```
